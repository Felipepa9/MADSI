```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/CIENCIA DE DATOS/datos.csv')

data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['total_pax_qty'] = data['total_pax_qty'].fillna(round(data['total_pax_qty'].mean(),1))
data['total_capacity_leg_qty'] = data['total_capacity_leg_qty'].fillna(round(data['total_capacity_leg_qty'].mean(),1))
data['fo_pct'] = data['fo_pct'].fillna(round(data['fo_pct'].mean(),1))
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['flight_leg_registration_code','actual_flight_leg_fleet_name','actual_flight_leg_fleet_name_code'], inplace=True)
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
correlation_matrix = data.corr(numeric_only=True)

plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
